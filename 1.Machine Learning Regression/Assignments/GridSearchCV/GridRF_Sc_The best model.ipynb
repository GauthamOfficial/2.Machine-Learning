```python
#importing the libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the Dataset
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
#Standardisation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#to calculate mean and variance
independent = sc.fit_transform(independent)
independent = sc.transform(independent)
```

```python
#GridSearch
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['squared_error', 'friedman_mse', 'poisson'], 'max_features':['auto', 'sqrt', 'log2'], 'n_estimators':[10, 100]}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose = 3, n_jobs = -1) 
#fitting the model for grid search
grid.fit(independent, dependent)
```

```python
#print best parameter after tuning
#print(grid.best_params_)
result = grid.cv_results_

print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(result)
```

```python
new_data = [[age_input, bmi_input, children_input, sex_male_input, smoker_yes_input]]
```

```python
new_data_standardized = sc.transform(new_data)
```

```python
Future_Prediction = grid.predict(new_data_standardized)
print("Future_Prediction={}".format(Future_Prediction))
```
