```python
#pandas is a Library and it contains the functions related to the database
import pandas as pd
```

```python
#calling the data set and assigned it to the dataset variable
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#Split Input
independent=dataset[["YearsExperience"]]
independent
```

```python
#Split Output
dependent=dataset[["Salary"]]
dependent
```

```python
#Split train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Model Creation or Learning
from sklearn.linear_model import LinearRegression #procedure
regressor=LinearRegression() #procedure
regressor.fit(X_train, y_train) 
#after executing this line there will be the weight(w) and the bias(b) found or The Model created.
```

```python
#checking the weight(w)
weight=regressor.coef_
weight
```

```python
#checking the bias(b)
bias=regressor.intercept_
bias
```

```python
#Evaluation Metrics
#for testing
y_pred=regressor.predict(X_test)
```

```python
#finding R^2 Value
from sklearn.metrics import r2_score
r_score=r2_score(y_test, y_pred)
```

```python
r_score
```

```python
#save the best model
import pickle
fileName="finalized_model_linear.sav"
#ctreating a file to save the model
```

```python
#save the model in to the created file
pickle.dump(regressor, open(fileName, 'wb')) #wb: write binary
```

```python
#Load the saved model
loaded_model=pickle.load(open("finalized_model_linear.sav", 'rb')) #rb: read binary
#input the value to be predict
result=loaded_model.predict([[int(input("Enter the years of Experience: "))]])
```
